```python
from bs4 import BeautifulSoup
import pandas as pd
import requests as rq

import networkx as nx
import statistics as sts
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

```python
dark_aca=rq.get("https://www.goodreads.com/shelf/show/dark-academia")
dark_aca.encoding="UTF-8"
academia_soup=BeautifulSoup(dark_aca.text)
```

```python
title=[]
author=[]
link=[]
tag=[]
rate=[]
num_rate=[]

#print(academia_soup.find_all("div", class_="elementList")[0].find("a", class_="bookTitle")["href"])


for i, item in enumerate(academia_soup.find_all("div", class_="elementList")):
    try:
        title.append(item.find("a", class_="bookTitle").text)
        link.append("goodreads.com"+item.find("a", class_="bookTitle")["href"])
        author.append(item.find("a", class_="authorName").text)
        small=item.find("span", class_="greyText smallText").text
        small=small.strip().split("—\n")
        
        rate.append(small[0].strip().split()[-1])
        num_rate.append(small[1].strip().split()[0])
    except:
        print(i)

aca_df=pd.DataFrame({"title":title, "author":author, "link":link,
                     "rating":rate, "num_rate":num_rate})
```
